```python
# Lets replace the null values from the city column by the maximum ties repeating number
dataset['city'] = dataset['city'].fillna(dataset['city'].mode()[0])
```

```python
dataset.city.isnull().sum()
```

```python
# replace the null values of the "winne" column by the respective "toss_winner" values.
dataset['winner'] = dataset['winner'].fillna(dataset['toss_winner'])
```

```python
dataset['winner'].isnull().sum()
```

```python
dataset.isnull().sum()
```

```python
# Lets replace inconsistent names first , like
# "MS Dhoni", "M.S. Dhoni", "Mahendra Singh Dhoni" → unify to "MS Dhoni".

dataset['player_of_match']  = dataset['player_of_match'].str.replace('M.S Dhoni','MS Dhoni')
```

```python
# Group other categories as the 'other' if they contain the null values
dataset['player_of_match'] = dataset['player_of_match'].fillna('other')
```

```python
# What is the win percentage of each team across all seasons?
teams = dataset['winner'].value_counts()
total_wins = dataset['winner'].count()
win_percentage = (teams/total_wins*100)
win_percentage
```
